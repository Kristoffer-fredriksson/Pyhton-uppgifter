### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python


#skriver ut x o y värde
x = np.linspace(-4, 4, 400)
y = x**2

# gör första
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("Object-Oriented Approach: y = x^2")
ax.set_xlabel("x")
ax.set_ylabel("y = x^2")


plt.show()

#gör andra
plt.plot(x, y)
plt.title("Pyplot Approach: y = x^2")
plt.xlabel("x")
plt.ylabel("y = x^2")

plt.show()
```
